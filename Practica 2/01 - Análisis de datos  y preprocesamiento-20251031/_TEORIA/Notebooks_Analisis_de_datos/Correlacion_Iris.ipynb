```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

**MATRIZ DE CORRELACION**

```python
datos = pd.read_csv(DATOS_DIR+'Iris.csv')

datos.iloc[:,:-1].corr()
```

```python
sns.heatmap(datos.iloc[:,:-1].corr(), square=True, annot=True)
```

**DIAGRAMA DE DISPERSION**

```python
plt.figure()
X = 'petallength'
Y = 'petalwidth'
plt.scatter(datos[X],datos[Y])
plt.xlabel( X )
plt.ylabel( Y )
plt.show()
```

```python
plt.figure()
X = 'sepallength'
Y = 'sepalwidth'
plt.scatter(datos[X],datos[Y])
plt.xlabel( X )
plt.ylabel( Y )
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.pairplot(datos.iloc[:,:-1])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(y=datos['sepalwidth'])
plt.title('Violin Plot of Sepal Width')
plt.ylabel('Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=datos['sepalwidth'])
plt.title('Box Plot of Sepal Width')
plt.ylabel('Sepal Width')
plt.show()
```
